```python
import yfinance as yf
import pandas as pd 
import numpy as np
#import statsmodels.api as sm
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None  # default='warn'

df  = yf.download ('^BVSP', 
                   period = "1y",
                   interval = "1d")
                   #progress = False)

df = df[['Open','Adj Close']]
df['Dif %'] = ''

for i in range (1, len(df)):
    calc2 = df['Adj Close'][i] / df['Adj Close'][i-1] 
    calc = (calc2 *100) - 100
    df['Dif %'][i] = calc
    
#df.tail()

#REMOVER A ULTIMA LINHA
df.drop(index=df.index[-1],axis=0,inplace=True)
df = df.tail(10)

MediaFec = df['Adj Close'].mean()

print(MediaFec)
display(df)
```
