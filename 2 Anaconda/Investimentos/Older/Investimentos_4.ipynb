```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web
from datetime import date
import plotly.graph_objects as go

#!pip install yfinance --upgrade --no-cache-dir
import yfinance as yf
yf.pdr_override()

tickers = [
            'ITUB4.SA',
            'PETR4.SA',
            'ITSA4.SA',
            'ABEV3.SA',
            'BBDC4.SA',
            'PETR3.SA',
]

#carteira = web.get_data_yahoo(tickers, period='1y')
carteira = web.get_data_yahoo(tickers, period='6mo')
carteira['Adj Close'].plot(figsize=(25,15), label=tickers)
plt.legend()
print(date.today())
```
