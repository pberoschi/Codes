```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from time import sleep
from selenium.webdriver.chrome.options import Options

%config IPCompleter.greedy=True

tempoBusca = 15

option = Options()
option.headless = True
navegador = webdriver.Chrome()                       #NAVEGADOR ON
#navegador = webdriver.Chrome(options=option)          #NAVEGADOR OFF
navegador.get('https://www.cartolafcbrasil.com.br/scouts')
sleep(1)

#---------------------- *** GOLEIROS *** ----------------------#
print('>>> BUSCANDO DADOS NO SITE... ((( GOLEIROS ))) <<<')

#posição: GOLEIROS
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_drpPosicoes"]/option[2]').click()
sleep(1)
#filtro: 100 itens
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_drpPageSize"]/option[3]').click()
sleep(1)
#clicar em filtrar
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_btnFiltrar"]').click()
sleep(tempoBusca)

tabela = navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_gvList"]')
html_content = tabela.get_attribute('outerHTML')
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find(name='table')

tabela_goleiro = pd.read_html(str(table),decimal=',',thousands='.')[0]

#substituir NaN por espaços em branco
tabela_goleiro = tabela_goleiro.fillna('')

table = soup.find(name='table')
listaBase = table.find_all('a',{'class':'mt5 pull-left'})

listatime = []
for x in listaBase:
    base = str(x['href'])
    time = base.split('/')[4]
    listatime.append(time.title())

time = pd.DataFrame(listatime)
#criando coluna 'time' e inserindo dados
tabela_goleiro.insert(2, 'Time', time)
#extraindo apenas as colunas necessárias
resumoGoleiros = tabela_goleiro[['Nome','Time','Preço','J','Média','Últ. Pont.','Variação']]
#organizando por descendente
FinalGoleiros = resumoGoleiros.sort_values(by=['J','Média','Últ. Pont.'],ascending=False)
#listando apenas os 10 primeiros
#print('\n\n''1) LISTA GERAL DE GOLEIROS')
#display(FinalGoleiros.head(20))



print('>>> BUSCANDO DADOS NO SITE... ((( LATERAIS ))) <<<')

#posição: LATERAIS
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_drpPosicoes"]/option[3]').click()
sleep(1)
#clicar em filtrar
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_btnFiltrar"]').click()
sleep(tempoBusca)

tabela = navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_gvList"]')
html_content = tabela.get_attribute('outerHTML')
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find(name='table')

tabela_lateral = pd.read_html(str(table),decimal=',',thousands='.')[0]

#substituir NaN por espaços em branco
tabela_lateral = tabela_lateral.fillna('')

table = soup.find(name='table')
listaBase = table.find_all('a',{'class':'mt5 pull-left'})

listatime = []
for x in listaBase:
    base = str(x['href'])
    time = base.split('/')[4]
    listatime.append(time.title())

time = pd.DataFrame(listatime)
#criando coluna 'time' e inserindo dados
tabela_lateral.insert(2, 'Time', time)
#extraindo apenas as colunas necessárias
resumoLaterais = tabela_lateral[['Nome','Time','Preço','J','Média','Últ. Pont.','Variação']]
#organizando por descendente
FinalLaterais = resumoLaterais.sort_values(by=['J','Média','Últ. Pont.'],ascending=False)
#listando apenas os 10 primeiros
#print('\n\n''1) LISTA GERAL DE LATERAIS')
#display(FinalLaterais.head(20))



print('>>> BUSCANDO DADOS NO SITE... ((( ZAGUEIROS ))) <<<')

#posição: ZAGUEIROS
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_drpPosicoes"]/option[4]').click()
sleep(1)
#clicar em filtrar
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_btnFiltrar"]').click()
sleep(tempoBusca)

tabela = navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_gvList"]')
html_content = tabela.get_attribute('outerHTML')
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find(name='table')

tabela_zagueiro = pd.read_html(str(table),decimal=',',thousands='.')[0]

#substituir NaN por espaços em branco
tabela_zagueiro = tabela_zagueiro.fillna('')

table = soup.find(name='table')
listaBase = table.find_all('a',{'class':'mt5 pull-left'})

listatime = []
for x in listaBase:
    base = str(x['href'])
    time = base.split('/')[4]
    listatime.append(time.title())

time = pd.DataFrame(listatime)
#criando coluna 'time' e inserindo dados
tabela_zagueiro.insert(2, 'Time', time)
#extraindo apenas as colunas necessárias
resumoZagueiros = tabela_zagueiro[['Nome','Time','Preço','J','Média','Últ. Pont.','Variação']]
#organizando por descendente
FinalZagueiros = resumoZagueiros.sort_values(by=['J','Média','Últ. Pont.'],ascending=False)
#listando apenas os 10 primeiros
#print('\n\n''1) LISTA GERAL DE ZAGUEIROS')
#display(FinalZagueiros.head(20))


print('>>> BUSCANDO DADOS NO SITE... ((( MEIAS ))) <<<')

#posição: MEIAS
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_drpPosicoes"]/option[5]').click()
sleep(1)
#clicar em filtrar
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_btnFiltrar"]').click()
sleep(tempoBusca)

tabela = navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_gvList"]')
html_content = tabela.get_attribute('outerHTML')
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find(name='table')

tabela_meia = pd.read_html(str(table),decimal=',',thousands='.')[0]

#substituir NaN por espaços em branco
tabela_meia = tabela_meia.fillna('')

table = soup.find(name='table')
listaBase = table.find_all('a',{'class':'mt5 pull-left'})

listatime = []
for x in listaBase:
    base = str(x['href'])
    time = base.split('/')[4]
    listatime.append(time.title())

time = pd.DataFrame(listatime)
#criando coluna 'time' e inserindo dados
tabela_meia.insert(2, 'Time', time)
#extraindo apenas as colunas necessárias
resumoMeias = tabela_meia[['Nome','Time','Preço','J','Média','Últ. Pont.','Variação']]
#organizando por descendente
FinalMeias = resumoMeias.sort_values(by=['J','Média','Últ. Pont.'],ascending=False)
#listando apenas os 10 primeiros
#print('\n\n''1) LISTA GERAL DE MEIAS')
#display(FinalMeias.head(20))



print('>>> BUSCANDO DADOS NO SITE... ((( ATACANTES ))) <<<')

#posição: ATACANTES
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_drpPosicoes"]/option[6]').click()
sleep(1)
#clicar em filtrar
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_btnFiltrar"]').click()
sleep(tempoBusca)

tabela = navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_gvList"]')
html_content = tabela.get_attribute('outerHTML')
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find(name='table')

tabela_atacante = pd.read_html(str(table),decimal=',',thousands='.')[0]

#substituir NaN por espaços em branco
tabela_atacante = tabela_atacante.fillna('')

table = soup.find(name='table')
listaBase = table.find_all('a',{'class':'mt5 pull-left'})

listatime = []
for x in listaBase:
    base = str(x['href'])
    time = base.split('/')[4]
    listatime.append(time.title())

time = pd.DataFrame(listatime)
#criando coluna 'time' e inserindo dados
tabela_atacante.insert(2, 'Time', time)
#extraindo apenas as colunas necessárias
resumoAtacantes = tabela_atacante[['Nome','Time','Preço','J','Média','Últ. Pont.','Variação']]
#organizando por descendente
FinalAtacantes = resumoAtacantes.sort_values(by=['J','Média','Últ. Pont.'],ascending=False)
#listando apenas os 10 primeiros
#print('\n\n''1) LISTA GERAL DE ATACANTES')
#display(FinalAtacantes.head(20))



print('>>> BUSCANDO DADOS NO SITE... ((( TÉCNICOS ))) <<<')

#posição: TÉCNICOS
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_drpPosicoes"]/option[7]').click()
sleep(1)
#clicar em filtrar
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_btnFiltrar"]').click()
sleep(tempoBusca)

tabela = navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_gvList"]')
html_content = tabela.get_attribute('outerHTML')
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find(name='table')

tabela_tecnico = pd.read_html(str(table),decimal=',',thousands='.')[0]

#substituir NaN por espaços em branco
tabela_tecnico = tabela_tecnico.fillna('')

table = soup.find(name='table')
listaBase = table.find_all('a',{'class':'mt5 pull-left'})

listatime = []
for x in listaBase:
    base = str(x['href'])
    time = base.split('/')[4]
    listatime.append(time.title())

time = pd.DataFrame(listatime)
#criando coluna 'time' e inserindo dados
tabela_tecnico.insert(2, 'Time', time)
#extraindo apenas as colunas necessárias
resumoTecnicos = tabela_tecnico[['Nome','Time','Preço','J','Média','Últ. Pont.','Variação']]
#organizando por descendente
FinalTecnicos = resumoTecnicos.sort_values(by=['J','Média','Últ. Pont.'],ascending=False)
#listando apenas os 10 primeiros
#print('\n\n''1) LISTA GERAL DE TECNICOS')
#display(FinalTecnicos.head(20))

print('\n\n DADOS COLETADOS COM SUCESSO...FECHANDO O NAVEGADOR')
navegador.close()
sleep(3)
print('\n\n PROCESSO FINALIZADO E DADOS DISPONÍVEIS')
```
