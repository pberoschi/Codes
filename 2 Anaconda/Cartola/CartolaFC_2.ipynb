```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from time import sleep
from selenium.webdriver.chrome.options import Options

%config IPCompleter.greedy=True

option = Options()
option.headless = True
#navegador = webdriver.Chrome()                       #NAVEGADOR ON
navegador = webdriver.Chrome(options=option)          #NAVEGADOR OFF
navegador.get('https://www.cartolafcbrasil.com.br/scouts')
sleep(10)
print('>>> BUSCANDO DADOS NO SITE... <<<')

#---------------------- *** GOLEIROS APTOS *** ----------------------#

#posição: GOLEIROS
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_drpPosicoes"]/option[2]').click()
sleep(1)
#filtro: 100 itens
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_drpPageSize"]/option[3]').click()
sleep(1)
#clicar em filtrar
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_btnFiltrar"]').click()
sleep(1)

tabela = navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_gvList"]')
html_content = tabela.get_attribute('outerHTML')
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find(name='table')

tabela_goleiro = pd.read_html(str(table),decimal=',',thousands='.')[0]

#substituir NaN por espaços em branco
tabela_goleiro = tabela_goleiro.fillna('')

testegoleiro = str(tabela_goleiro.iloc[0:1,1])[5:39]

if testegoleiro != 'Não existem registros cadastrados!':
    table = soup.find(name='table')
    listaBase = table.find_all('a',{'class':'mt5 pull-left'})

    listatime = []
    for x in listaBase:
        base = str(x['href'])
        time = base.split('/')[4]
        listatime.append(time.title())
        
    time = pd.DataFrame(listatime)
    #criando coluna 'time' e inserindo dados
    tabela_goleiro.insert(2, 'Time', time)
    #extraindo apenas as colunas necessárias
    resumoGoleiros = tabela_goleiro[['Nome','Time','Preço','J','Média','Últ. Pont.','Variação']]
    #organizando por descendente
    FinalGoleiros = resumoGoleiros.sort_values(by=['J','Média','Últ. Pont.'],ascending=False)
    #listando apenas os 10 primeiros
    print('\n\n''1) LISTA GERAL DE GOLEIROS')
    display(FinalGoleiros)
else:
    print('\n\n''1) >>>''Não existem cadastros de goleiros aptos!''<<<'.upper())


#---------------------- *** GOLEIROS SUSPENSOS *** ----------------------#

#condição: SUSPENSO
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_drpStatus"]/option[3]').click()
sleep(1)
#clicar em filtrar
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_btnFiltrar"]').click()
sleep(1)

tabela = navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_gvList"]')
html_content = tabela.get_attribute('outerHTML')
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find(name='table')
tabela = navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_gvList"]')
html_content = tabela.get_attribute('outerHTML')
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find(name='table')

tabela_goleiroSusp = pd.read_html(str(table),decimal=',',thousands='.')[0]

#substituir NaN por espaços em branco
tabela_goleiroSusp = tabela_goleiroSusp.fillna('')

#teste: acrescentar informação
#tabela_goleiroSusp.insert(1, 'Time', 'teste')
#display(tabela_goleiroSusp)

testegoleiroSusp = str(tabela_goleiroSusp.iloc[0:1,1])[5:39]

if testegoleiroSusp != 'Não existem registros cadastrados!':
    table = soup.find(name='table')
    listaBase = table.find_all('a',{'class':'mt5 pull-left'})
    listatimeSusp = []
    for x in listaBase:
        base = str(x['href'])
        time = base.split('/')[4]
        listatimeSusp.append(time.title())
        
    time = pd.DataFrame(listatimeSusp)
    
    #criando coluna 'time' e inserindo dados
    tabela_goleiroSusp.insert(2, 'Time', time)
    #extraindo apenas as colunas necessárias
    resumoGoleirosSusp = tabela_goleiroSusp[['Nome','Time','Preço','J','Média','Últ. Pont.','Variação']]
    #organizando por descendente
    FinalGoleirosSusp = resumoGoleirosSusp.sort_values(by=['J','Média','Últ. Pont.'],ascending=False)
    print('\n\n''2) LISTA GERAL DE GOLEIROS SUSPENSOS')
    display(FinalGoleirosSusp)
else:
    print('\n\n''2) >>>'' Não existem cadastros de goleiros suspensos! ''<<<'.upper())
    
#---------------------- *** GOLEIROS CONTUNDIDOS *** ----------------------#
    
#Status: CONTUNDIDOS
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_drpStatus"]/option[4]').click()
#Filtrar
navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_btnFiltrar"]').click()
sleep(1)

tabela = navegador.find_element_by_xpath('//*[@id="ctl00_cphMainContent_gvList"]')
html_content = tabela.get_attribute('outerHTML')
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find(name='table')
tabela_goleiroContund = pd.read_html(str(table),decimal=',',thousands='.')[0]

#substituir NaN por espaços em branco
tabela_goleiroContund = tabela_goleiroContund.fillna('')

testegoleiroContund = str(tabela_goleiroContund.iloc[0:1,1])[5:39]

if testegoleiroContund != 'Não existem registros cadastrados!':
    table = soup.find(name='table')
    listaBase = table.find_all('a',{'class':'mt5 pull-left'})

    listatimeContund = []
    for x in listaBase:
        base = str(x['href'])
        time = base.split('/')[4]

        listatimeContund.append(time.title())
        #print(time.title())

    time = pd.DataFrame(listatimeContund)

    #criando coluna 'time' e inserindo dados
    tabela_goleiroContund.insert(2, 'Time', time)
    #extraindo apenas as colunas necessárias
    resumoGoleirosContund = tabela_goleiroContund[['Nome','Time','Preço','J','Média','Últ. Pont.','Variação']]
    #organizando por descendente
    FinalGoleirosContund = resumoGoleirosContund.sort_values(by=['J','Média','Últ. Pont.'],ascending=False)
    #listando goleiros contundidos
    print('\n\n''3) LISTA GERAL DE GOLEIROS CONTUNDIDOS')
    display(FinalGoleirosContund.head())
else:
    print('\n\n''3) >>>'' Não existem cadastros de goleiros contundidos! ''<<<'.upper())
     
```
